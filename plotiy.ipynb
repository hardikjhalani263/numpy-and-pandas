```python
# Employee attrition analysis
# attrition -----> employes , company leave.
# quaterly basis , years basis

# Employes attriation analysis is a type of behavioural analysis where we study the behaviour and characteristics of
# the employees who left the organization and compare theis characteristics with the current employees to find the
# employees who may leave the organization soon.

# A high rate of attrition of employees can be expensive for any company in terms of recruitment and traning costs
# , loss of productivity and morale reduction of employees . by identifying the causes of attrition , a company
# can taken measures to reduce the attrition of employees and maintain precious employees.
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Attrition -project(2).csv')
df.head(2)
```

```python
df.columns
```

```python
df.shape
```

```python
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
# filter the data to show only "yes " values in the "attrition " column
```

```python
attr_df = df [df['Attrition'] == 'Yes']
attr_df
```

```python
# Department wise Attrition
```

```python
atr_by_dpt = attr_df.groupby('Department').size().reset_index(name='count')
```

```python
atr_by_dpt
```

```python
# creat a donut chart
```

```python
fig = go.Figure(data=[go.Pie(
labels=atr_by_dpt['Department'],
values=atr_by_dpt['count'],
hole=0.4,
marker=dict(colors=['red','green']),
textposition='inside'
)])
# update the layout
fig.update_layout(title='Employee attrition by department')

# show the chart
fig.show()
```
